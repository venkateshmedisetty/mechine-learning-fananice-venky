```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample dataset with at least 6 numerical columns
np.random.seed(42)
df = pd.DataFrame(np.random.rand(100, 6), columns=[
    'feature_0', 'feature_1', 'feature_2',
    'feature_3', 'feature_4', 'feature_5'
])

# Step 3: Determine selected columns based on SID 2281544
col1_index = 4
col2_index = 5
col1 = df.columns[col1_index]
col2 = df.columns[col2_index]

# Step 4: Plotting the bicolour interaction diagram
plt.figure(figsize=(8, 6))
sns.kdeplot(
    x=df[col1],
    y=df[col2],
    cmap="coolwarm",
    fill=True,
    thresh=0,
    levels=100
)
plt.title(f'Bicolour Feature Interaction: {col1} vs {col2}')
plt.xlabel(col1)
plt.ylabel(col2)
plt.grid(True)
plt.tight_layout()
plt.show()
```
