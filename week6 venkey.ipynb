```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode

# Enable Plotly in offline mode
init_notebook_mode(connected=True)

# Create dummy dataset with High_Bid and Low_Bid
np.random.seed(42)
rows = 90000  # Make sure it's more than 81544 + 544
df = pd.DataFrame({
    'High_Bid': np.random.uniform(100, 200, size=rows),
    'Low_Bid': np.random.uniform(80, 100, size=rows)
})

# SID 2281544 → Start at 81544, for 544 rows
start_index = 81544
time_period = 544
end_index = start_index + time_period

# Extract the subset
subset = df.iloc[start_index:end_index]

# Create traces
trace1 = go.Scatter(x=subset.index, y=subset['High_Bid'], mode='lines', name='High Bid', line=dict(color='green'))
trace2 = go.Scatter(x=subset.index, y=subset['Low_Bid'], mode='lines', name='Low Bid', line=dict(color='red'))

# Plot
layout = go.Layout(title='High_Bid vs Low_Bid', xaxis=dict(title='Index'), yaxis=dict(title='Price'))
fig = go.Figure(data=[trace1, trace2], layout=layout)
iplot(fig)
```
